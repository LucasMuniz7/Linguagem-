## 4. Missão 1 — Barras: Receita por Canal

A missão prática da aula pede construir um gráfico de barras para responder:
**qual canal vende mais?**  
Os slides também sugerem usar o hover para injetar métricas secundárias sem poluir a tela. fileciteturn7file0

### Tarefa
1. Agregue a `receita` por `canal_venda`
2. Inclua também uma métrica secundária, como `quantidade`
3. Construa um gráfico de barras com Plotly Express
4. Ordene do maior para o menor
5. Faça o hover mostrar mais do que apenas a receita

### Perguntas
- Qual canal lidera a receita?
- O canal líder também lidera em quantidade?
- O hover ajudou a enriquecer a leitura sem poluir a tela?


```python
canal = df.groupby('canal_venda').agg(
    receita_total=('receita', 'sum'),
    quantidade_total=('quantidade', 'sum')
).reset_index().sort_values('receita_total', ascending=False)

fig_barras_canal = px.bar(
    canal,
    x='canal_venda',
    y='receita_total',
    hover_data={'quantidade_total': True, 'receita_total': ':,.2f'},
    title='Canal Online lidera receita — mas qual tem melhor eficiência por venda?',
    labels={
        'canal_venda': 'Canal de Venda',
        'receita_total': 'Receita Total (R$)',
        'quantidade_total': 'Qtd. Vendida'
    },
    color='canal_venda',
    text_auto=True
)

fig_barras_canal.update_layout(showlegend=False)
fig_barras_canal.show()
```

### Insight obrigatório
Escreva 2 ou 3 linhas explicando:
- quem lidera
- quem fica atrás
- o que um gestor poderia investigar a seguir


O canal Online concentra o maior volume de receita, enquanto os demais ficam atrás
em volume total. O hover revela a quantidade vendida sem poluir o gráfico, permitindo
que o gestor investigue se o canal líder também vende mais unidades ou apenas em
tickets maiores — e decida onde concentrar esforços de crescimento.


## 5. Missão 2 — Linhas: Sazonalidade da Receita Mensal

Os slides destacam que o gráfico de linhas, com zoom e pan, é ideal para navegar no tempo e isolar picos. Também reforçam que o eixo temporal precisa estar corretamente tipado no Pandas. fileciteturn7file0

### Tarefa
1. Converta `data_venda` em data, se necessário
2. Agregue a `receita` por mês
3. Construa um gráfico de linha interativo
4. Teste o zoom nos meses de pico
5. Observe se novembro e dezembro se destacam

### Perguntas
- Existe sazonalidade?
- Quais meses chamam atenção?
- O zoom ajudou a explorar melhor a parte final da série?


```python
# Construa aqui o gráfico de linha interativo da receita mensal
df['mes'] = df['data_venda'].dt.to_period('M').dt.to_timestamp()

mensal = df.groupby('mes')['receita'].sum().reset_index()
mensal.columns = ['mes', 'receita_total']

fig_linha_mensal = px.line(
    mensal,
    x='mes',
    y='receita_total',
    title='Receita acelera no segundo semestre e atinge pico em novembro e dezembro',
    labels={
        'mes': 'Mês',
        'receita_total': 'Receita Total (R$)'
    },
    markers=True
)

fig_linha_mensal.update_layout(
    xaxis_title='Mês',
    yaxis_title='Receita (R$)',
    hovermode='x unified'
)

fig_linha_mensal.show()
```

### Insight obrigatório
Explique:
- qual tendência aparece
- se há picos claros
- qual hipótese de negócio pode explicar o comportamento observado


Existe sazonalidade clara com pico no fim do ano, padrão típico do varejo brasileiro
(Black Friday e Natal). O zoom interativo permite isolar esses meses e comparar com
o mesmo período do ano anterior. A hipótese de negócio é que campanhas sazonais
concentram receita — o que justifica planejamento de estoque e logística antecipado.


## 6. Missão 3 — Dispersão: Lucro vs. Receita segmentado por Categoria

Os slides mostram o uso do gráfico de dispersão para revelar correlação entre KPIs e identificar anomalias, com grande apoio do hover. fileciteturn7file0

### Tarefa
1. Construa um scatter plot com:
   - eixo X = `receita`
   - eixo Y = `lucro`
2. Use `categoria` como cor
3. Inclua no hover:
   - `produto`
   - `canal_venda`
   - `uf`
   - `margem_lucro`
4. Tente identificar algum ponto fora da curva

### Perguntas
- Existe correlação visual entre receita e lucro?
- Há anomalias?
- O hover ajuda a transformar um ponto em um caso investigável?


```python
# Construa aqui o scatter plot interativo entre receita e lucro
fig_scatter = px.scatter(
    df,
    x='receita',
    y='lucro',
    color='categoria',
    hover_data={
        'produto': True,
        'canal_venda': True,
        'uf': True,
        'margem_lucro': ':.2%'
    },
    title='Alta receita nem sempre significa alto lucro — anomalias merecem investigação',
    labels={
        'receita': 'Receita (R$)',
        'lucro': 'Lucro (R$)',
        'categoria': 'Categoria'
    },
    opacity=0.6
)

fig_scatter.show()
```

### Insight obrigatório
Escreva 2 ou 3 linhas dizendo:
- se a correlação parece positiva
- onde surgem pontos fora do padrão
- que ação analítica poderia ser tomada depois


A correlação entre receita e lucro é positiva na maioria dos pontos, mas existem
casos com receita alta e lucro baixo ou negativo — possíveis anomalias de custo,
desconto excessivo ou devolução. O hover transforma esses pontos suspeitos em casos
investigáveis, revelando produto, canal e UF de cada ocorrência para que o gestor
tome ação direta.


## 7. Exploração espacial — Onde está concentrada a operação?

Os slides incluem mapas geográficos como resposta para perguntas espaciais como:
**onde está concentrada nossa operação?** fileciteturn7file0

### Tarefa
Usando `latitude` e `longitude`, crie um mapa interativo que ajude a visualizar a operação.

### Sugestões
- use tamanho ou cor para uma métrica relevante (como receita)
- teste o zoom do mapa
- observe se há concentração regional

### Perguntas
- Onde a operação parece mais concentrada?
- O mapa ajudou mais do que uma simples tabela por UF?


```python
# Construa aqui um mapa interativo com Plotly
uf_receita = df.groupby(['uf', 'latitude', 'longitude']).agg(
    receita_total=('receita', 'sum')
).reset_index()

fig_mapa = px.scatter_geo(
    uf_receita,
    lat='latitude',
    lon='longitude',
    size='receita_total',
    hover_name='uf',
    hover_data={'receita_total': ':,.2f', 'latitude': False, 'longitude': False},
    title='Sudeste concentra a maior parte da operação — Sul e Nordeste emergem',
    scope='south america',
    size_max=40
)

fig_mapa.show()
```
