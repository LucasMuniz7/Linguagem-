## 4. Batalha 1 — A tirania dos tipos de dados

Nos slides, vimos que datas não podem ser tratadas como texto e que números em formato string quebram análises. fileciteturn2file0

### Tarefa
Converta corretamente, quando fizer sentido:
- `data`
- `receita`
- `lucro`
- `quantidade`

### Orientação
- Investigue valores estranhos antes de converter
- Pense no uso de `errors='coerce'`
- Registre em markdown ou comentário quais problemas encontrou


```python
# 1. Investigar e converter a coluna 'data'
# Tentar converter formatos comuns (YYYY-MM-DD, YYYY/MM/DD) sem dayfirst=True primeiro
df['data'] = pd.to_datetime(df['data'], errors='coerce')

# Para os valores NaT restantes, tentar converter DD/MM/YYYY usando dayfirst=True
mask_nat_dates = df['data'].isna()
df.loc[mask_nat_dates, 'data'] = pd.to_datetime(df.loc[mask_nat_dates, 'data'], errors='coerce', dayfirst=True)

print("Valores únicos na coluna 'data' após a conversão refinada (amostra):\n", df['data'].unique()[:10])
print("\nContagem de valores nulos na coluna 'data' após a conversão refinada:\n", df['data'].isnull().sum())

# 2. Investigar e converter a coluna 'receita'
# Limpar a coluna 'receita': remover 'R$', substituir '.' (separador de milhares) e ',' (separador decimal), remover espaços e converter 'zero' para 0
df['receita'] = df['receita'].astype(str).str.replace('R$', '', regex=False)
df['receita'] = df['receita'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).str.strip()
df['receita'] = df['receita'].replace('zero', '0') # Substituir a string 'zero' pelo número 0
df['receita'] = pd.to_numeric(df['receita'], errors='coerce')
print("\nContagem de valores nulos na coluna 'receita' após a conversão refinada:\n", df['receita'].isnull().sum())

# 3. Converter a coluna 'lucro' (já é float, mas aplicar coerce para consistência e segurança)
df['lucro'] = pd.to_numeric(df['lucro'], errors='coerce')
print("\nContagem de valores nulos na coluna 'lucro' após a conversão:\n", df['lucro'].isnull().sum())

# 4. Coluna 'quantidade' já é int64, não precisa de conversão.

# Confirmar os dtypes finais
print("\nTipos de dados após as conversões refinadas:\n")
df.info()
```

## 5. Batalha 2 — O enigma dos valores ausentes

A aula reforça que valores ausentes não devem ser tratados automaticamente; a decisão precisa ser justificada. fileciteturn2file0

### Tarefa
1. Mapeie os valores ausentes por coluna
2. Identifique quais colunas críticas têm nulos
3. Defina uma estratégia para cada caso:
   - remover linhas?
   - preencher?
   - manter?
4. Justifique cada escolha

### Dica
Evite decisões automáticas sem análise de contexto.


```python
print("Contagem de valores nulos por coluna (após conversões de tipo):\n", df.isnull().sum())


for col in ['uf', 'canal', 'categoria']:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)

df.dropna(subset=['receita', 'lucro'], inplace=True)

df['observacao'].fillna('Sem Observação', inplace=True)

print("\nContagem de valores nulos por coluna (após tratamento):\n", df.isnull().sum())

df.reset_index(drop=True, inplace=True)
```

### Registro reflexivo sobre o tratamento de nulos:

*   **Colunas em que removi linhas:**
    *   `receita` e `lucro`: Foram removidas as linhas onde estas colunas possuíam valores nulos. A justificativa é que `receita` e `lucro` são métricas financeiras críticas para qualquer análise de negócio. Imputar valores poderia distorcer significativamente os KPIs e as conclusões. Remover as poucas linhas com nulos (15 para `receita` e 6 para `lucro`) garante a integridade dos cálculos financeiros.

*   **Colunas em que preenchi valores:**
    *   `uf`, `canal`, `categoria`: Nestas colunas categóricas (com 5 nulos cada), optei por preencher os valores ausentes com a moda (o valor mais frequente) de cada coluna. A quantidade de nulos era pequena, e preencher com a moda ajuda a preservar as observações restantes no dataset, além de manter a distribuição existente da categoria sem adicionar uma nova categoria 'Desconhecido' que poderia não ser útil para análise.
    *   `observacao`: Esta coluna, sendo de natureza descritiva e não crítica para métricas numéricas, teve seus valores nulos preenchidos com a string 'Sem Observação'. Isso torna explícita a ausência de uma observação sem a necessidade de remover a linha inteira, que poderia conter dados úteis em outras colunas.

*   **Situações em que decidi manter o nulo:**
    *   `data`: A coluna `data` apresentou um número extremamente alto de valores nulos (224 de 228) após as tentativas de conversão. Isso indica um problema fundamental na formatação dos dados originais que não pôde ser resolvido com as abordagens de conversão padrão do Pandas. Remover praticamente todas as linhas inviabilizaria a atividade. Para fins práticos desta atividade, decidi manter os valores `NaT` na coluna `data`, reconhecendo que isso impõe uma severa limitação para qualquer análise temporal. Em um cenário real, uma investigação aprofundada na fonte de dados original seria essencial para entender e corrigir a raiz do problema de formatação das datas, e estas linhas seriam tratadas como inválidas para análises que dependessem de datas válidas.

## 6. Batalha 3 — O ataque dos clones (duplicidades)

A aula alerta que nem toda duplicidade é erro automático: pode haver compra legítima repetida ou dupla inserção do sistema. fileciteturn2file0

### Tarefa
1. Investigue se há linhas duplicadas
2. Analise se a duplicidade parece nociva para os KPIs
3. Escolha uma estratégia:
   - remover duplicidades totais?
   - remover apenas com base em certas colunas?
   - manter?
4. Justifique a decisão


```python
# 1. Investigar se há linhas duplicadas
duplicatas_totais = df.duplicated().sum()
print(f"Número de linhas duplicadas (exatas): {duplicatas_totais}")

# Exibir as linhas duplicadas, se houver, para análise
if duplicatas_totais > 0:
    print("\nExemplo de linhas duplicadas:")
    display(df[df.duplicated(keep=False)].sort_values(by=list(df.columns)).head(10)) # Exibe as duplicadas e suas originais

# 2. Análise e escolha de estratégia
# Assumindo que linhas totalmente duplicadas são erros de registro, elas devem ser removidas.
# Se houvesse duplicidade apenas em 'pedido_id' mas com outras colunas diferentes, isso indicaria
# um problema mais complexo na chave primária ou múltiplas entradas para o mesmo pedido, o que exigiria outra abordagem.
# Para este dataset, a duplicidade total é o cenário mais provável de erro de entrada.

# 3. Remover duplicidades totais
df.drop_duplicates(inplace=True)
print(f"\nNúmero de linhas após remover duplicatas: {len(df)}")
```

## 7. Padronização de categorias

Os slides mostram que categorias duplicadas ou mal escritas podem distorcer rankings e filtros. fileciteturn2file0

### Tarefa
Inspecione e padronize, se necessário:
- `uf`
- `canal`
- `categoria`

### Pense em:
- maiúsculas e minúsculas
- espaços extras
- acentuação / variações
- categorias equivalentes escritas de formas diferentes


```python
df['uf'] = df['uf'].str.upper().str.strip()
df['canal'] = df['canal'].str.title().str.strip().replace({'Loja Fisica': 'Loja Física'})
df['categoria'] = df['categoria'].str.title().str.strip().replace({'Audio': 'Áudio', 'Informatica': 'Informática', 'Perifericos': 'Periféricos'})
```

## 8. Subindo de nível — Criação de variáveis derivadas

Depois da limpeza, é hora de enriquecer a base com variáveis úteis para análise. fileciteturn2file0

### Tarefa
Crie, se possível:
- `ano`
- `mes`
- `ano_mes`
- `margem_lucro`

### Atenção
A criação de `margem_lucro` exige cuidado com divisões por zero e possíveis valores infinitos.


```python
df['ano'] = df['data'].dt.year
df['mes'] = df['data'].dt.month
df['ano_mes'] = df['data'].dt.strftime('%Y-%m')

df['margem_lucro'] = df.apply(lambda row: row['lucro'] / row['receita'] if row['receita'] != 0 else 0, axis=1)
```
